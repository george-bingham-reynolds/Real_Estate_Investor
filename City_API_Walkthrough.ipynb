```python
import requests
import pandas as pd
import json
import http.client
```

```python
# CHECK THAT WE SUCCESSFULY HIT API
r = requests.get('http://127.0.0.1:5000/city')
r.status_code
```

```python
# CHECK THAT CITY LIST IS CURRENTLY EMPTY
json.loads(r.text)
```

```python
# READ IN DATAFRAME OF CITY INFO W/ PREDICTIONS
df = pd.read_csv("city_level_files/latest_yrs_w_preds_city.csv")

# CURRENT API SETUP CAN'T HANDLE THE ACCENT MARKS, AND THE '/' BREAKS URL FORMAT
# FUTURE ITERATIONS SHOULD INCLUDE A FIX
df = df[~df['place'].isin(['Mayagüez.zona.urbana.Puerto.Rico', 
                             'Bayamón.zona.urbana.Puerto.Rico',
                             'Louisville/Jefferson.County.metro.government.(balance).Kentucky'])].reset_index(drop = True)
```

BELOW USES POST METHOD TO ADD DATAFRAME OF CITY INFO AND ROI PREDICTIONS TO API

```python
# LOOP THRU AND POST CITY INFO TO API

for i in range(len(df)):
    
    # CONVERT ROW TO DICTIONARY TO PASS THRU
    iter_record = df.loc[i].to_dict()

    
    # DEFINE CONNECTION, PARAMETERS
    conn = http.client.HTTPConnection("127.0.0.1:5000")
    payload = json.dumps(iter_record)
    headers = {
        'Content-Type': "application/json",
        'User-Agent': "insomnia/8.6.1"
        }

    # SEND POST REQUEST
    conn.request("POST", "/city", payload, headers)
```

```python
# CHECK THAT WE SUCCESSFULY POSTED INFO
r = requests.get('http://127.0.0.1:5000/city')
json.loads(r.text)
```
