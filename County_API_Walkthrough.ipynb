```python
import requests
import pandas as pd
import json
import http.client
```

```python
# CHECK THAT WE SUCCESSFULY HIT API
r = requests.get('http://127.0.0.1:5000/county')
r.status_code
```

```python
# CHECK THAT COUNTY LIST IS CURRENTLY EMPTY
json.loads(r.text)
```

```python
# READ IN DATAFRAME OF COUNTY INFO W/ PREDICTIONS
df = pd.read_csv("county_level_files/latest_yrs_w_preds_county.csv")

# CURRENT API SETUP CAN'T HANDLE THE ACCENT MARKS - FUTURE ITERATIONS SHOULD INCLUDE A FIX
df = df[~df['place'].isin(['Mayagüez.Municipio.Puerto.Rico', 
                             'Bayamón.Municipio.Puerto.Rico'])].reset_index(drop = True)
```

BELOW USES POST METHOD TO ADD DATAFRAME OF CITY INFO AND ROI PREDICTIONS TO API

```python
# LOOP THRU AND POST COUNTY INFO TO API

for i in range(len(df)):
    
    # CONVERT ROW TO DICTIONARY TO PASS THRU
    iter_record = df.loc[i].to_dict()

    
    # DEFINE CONNECTION, PARAMETERS
    conn = http.client.HTTPConnection("127.0.0.1:5000")
    payload = json.dumps(iter_record)
    headers = {
        'Content-Type': "application/json",
        'User-Agent': "insomnia/8.6.1"
        }

    # SEND POST REQUEST
    conn.request("POST", "/county", payload, headers)
```

```python
# CHECK THAT WE SUCCESSFULY POSTED INFO
r = requests.get('http://127.0.0.1:5000/county')
json.loads(r.text)[:10]
```

LASTLY, RUN THE BELOW TO DELETE ALL DATA FROM THE API IF YOU WANT IT CLEARED

```python
conn = http.client.HTTPConnection("127.0.0.1:5000")

payload = ""

headers = { 'User-Agent': "insomnia/8.6.1" }

conn.request("DELETE", "/county", payload, headers)
```
